```python
#scatter plot (survival vs age)
fig, axis = plt.subplots(1,1,figsize=(6,3))
axis.scatter(titanic.Age, titanic['Survived'], s=30, alpha =0.25, color ='pink')

#axis.set_ylim([0,4000000])
plt.show()
```

```python
titanic_survival = titanic.groupby(['Sex','Pclass'])['Survived'].sum()
titanic_survival = pd.DataFrame(titanic_survival)
titanic_survival

```

```python

sns.barplot(data=titanic, x='Sex', y='Survived',estimator = sum, hue='Pclass',palette =sns.color_palette("colorblind", 10))
```

```python
titanic_survival = titanic.groupby(['Sex','Pclass','Age'])['Survived'].sum()
titanic_survival = pd.DataFrame(titanic_survival)
titanic_survival = titanic_survival.reset_index()
titanic_survival1 = titanic_survival[titanic_survival['Pclass']== 1]
titanic_survival1
titanic_survival2 = titanic_survival[titanic_survival['Pclass']== 2]
titanic_survival3 = titanic_survival[titanic_survival['Pclass']== 3]
titanic_survival2
```

```python
frames = [titanic_survival1,titanic_survival2,titanic_survival3]
results = pd.concat(frames)
results
```

```python
fig, ax= plt.subplots(1,1, figsize=(10,5))
sns.set(font_scale = 1)


sns.regplot(x="Age", y="Survived", data=titanic_survival1, ci = False, 
    scatter_kws={"color":"darkred","alpha":0.3,"s":90},
    line_kws={"color":"darkred","alpha":0.9,"lw":4},marker="1", ax=ax, label='Class 1')

sns.regplot(x="Age", y="Survived", data=titanic_survival2, ci = False, 
    scatter_kws={"color":"green","alpha":0.3,"s":90},
    line_kws={"color":"green","alpha":0.9,"lw":4},marker="^", ax=ax, label='Class 2')

sns.regplot(x="Age", y="Survived", data=titanic_survival3, ci = False, 
    scatter_kws={"color":"darkblue","alpha":0.3,"s":90},
    line_kws={"color":"darkblue","alpha":0.9,"lw":4},marker="8", ax=ax, label='Class 3')



# remove the top and right line in graph
#sns.despine()

ax.legend(loc='upper right')
ax.set_xlabel("Age",color="black",alpha=0.9)
ax.set_ylabel("Survived",color="black",alpha=0.9)

# Set the size of the graph from here
# g.figure.set_size_inches(12,8
# # Set the Title of the graph from here
# g.axes.set_title('Total Bill vs. Tip', fontsize=34,color="r",alpha=0.5)
# # Set the xlabel of the graph from here
# g.set_xlabel("Tip",size = 67,color="r",alpha=0.5)
# # Set the ylabel of the graph from here
# #g.set_ylabel("Total Bill",size = 67,color="r",alpha=0.5)
# # Set the ticklabel size and color of the graph from here
# g.tick_params(labelsize=14,labelcolor="black")
```
